```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("read_hive_table").enableHiveSupport().getOrCreate()
```

```python
query = """
SELECT *
FROM ks_ad_antispam.refund_routine_wide_for_policy_di
WHERE dt >= '20260101' AND dt <= '20260201'
"""

df = spark.sql(query)
```

```python
print(f"总行数: {df.count()}")
print(f"总列数: {len(df.columns)}")
df.printSchema()
```
